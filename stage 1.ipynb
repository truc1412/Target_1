```python
'''Bài 1: Viết chương trình nhập vào 1 ngày (ngày, tháng, năm). Tìm ngày kế ngày vừa nhập (ngày, tháng, năm)'''
day=int(input("Nhập ngày: "))
month=int(input("Nhập tháng: "))
year= int(input("Nhập năm: "))
if (month == 1) or (month == 3) or (month == 5) or (month == 7) or (month == 8) or (month == 10) or (month == 12):
    if day < 31:
        day+=1
    elif day == 31:
        if month == 12:
            year += 1
            day=1
            month=1
        else:
            day=1
            month += 1
elif (month == 4) or (month == 6) or (month == 9) or (month == 11):
    if day < 30:
        day+=1
    elif day == 30:
        day=1
        month+=1
elif month==2:
    if (year % 400 == 0) or ((year % 4 == 0) and (year % 100 != 0)):
        if day < 29:
            day+=1
        if day == 29:
            day=1
            month+=1
    else:
        if day < 28:
            day+=1
        if day == 28:
            day=1
            month+=1
print(f"Ngày kế tiếp là: {day}/{month}/{year}")
```

```python
"""Bài 2: Cho 1 Chuỗi như sau “5;7;8;-2;8;11;13;9;10” (có thể nhập bất kỳ từ bàn phím)- 
xuất các chữ số trên các dòng riêng biệt- Xuất có bao nhiêu chữ số chẵn- Xuất 
có bao nhiêu số âm- Xuất có bao nhiêu chữ số nguyên tố- Tính giá trị trung bình"""
chuoi=input("Nhập chuỗi số: ")
x=0
y=0
z=0
k=0
a=chuoi.split(";")
t=len(a)
for i in a:
    print(i)
    z=z+int(i)
tb=z/t    
for j in a:
    if int(j) % 2 == 0:
        x+=1
    if int(j) < 0:
        y+=1
    m=True
    if int(j) <2:
        m=False
    elif int(j) == 2:
        m=True
    else:
        for r in range (2,int(j)):
            if(int(j)%r ==0):
                m=False
    if m==True:
        k+=1

print(f"Có {x} chữ số chẵn")
print(f"Có {y} chữ số âm")
print(f"Có {k} chữ số nguyên tố")
print(f"Giá trị trung bình là: {tb}")
```

```python
"""Bài 3: Viết chương trình nhập vào độ dài 3 cạnh của một tam giác a, b, c.
Độ dài ba cạnh trên có pải là độ dài của một tam giác không?
- Kiểm tra tam giác có phải tam giác vuông không?
- Kiểm tra tam giác có phải tam giác đều không?
- Tính diện tích của tam giác dựa trên công thức sau:
"""
a=int(input("Nhập cạnh a: "))
b=int(input("Nhập cạnh b: "))
c=int(input("Nhập cạnh c: "))
if (a+b > c) and (b+c > a) and (a+c > b):
    print("Độ dài ba cạnh là độ dài của một tam giác")
    if(a*a == b*b + c*c or b*b == a*a + c*c or c*c == a*a + b*b):
        print("Đây là tam giác vuông")
    else:
        print("Đây không phải tam giác vuông")
    if (a==b and b==c):
        print("Đây là tam giác đều")
    else:
        print("Đây không phải tam giác đều")
    s=((a+b+c)*(b+c-a)*(a-b+c)*(a+b-c))/8
    print(f"Tam giác có diện tích là: {s}");
else:
    print("Độ dài ba cạnh không phải là độ dài của một tam giác")
```
